```
#Start by adding the necessary libraries
import pandas as pd
import numpy as np
```

```
# add datasets into environment
df1 = pd.read_csv('http://bit.ly/Telcom_dataset1')
df1.head()
```

```
df2 = pd.read_csv('http://bit.ly/Telcom_dataset2')
df2.head()
```

```
df3 = pd.read_csv('http://bit.ly/Telcom_dataset3')
df3.head()
```

```
#download the csv file and upload 
df4= pd.read_csv('cells_geo.csv', delimiter=';')
df4.head()
```

```
#Data Cleaning
#Make column names similar in df1, df2 and df3
df1.rename(columns = {'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME'}, inplace = True)
print(df1.columns)
```

```
df2.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace = True) 
print(df2.columns)
```

```
df3.rename(columns = {'CELLID':'CELL_ID', 'SIET_ID': 'SITE_ID'}, inplace = True) 
print(df3.columns)
```

```
df4.rename(columns = {'SITE_CODE': 'SITE_ID'}, inplace = True) 
print(df4.columns)
```

```
#Merge the above
df_all = [df1, df2, df3]
df_telcom = pd.concat(df_all)

df_telcom.tail()
```

```
df_n = df_telcom.merge(df4, left_on="SITE_ID", right_on ="SITE_ID")
df_n
```

```
#Checking for any missing values in the merged data set

df_n.isnull().sum()
```

```
# We fill the missing values with 0

df_n.fillna(value='0', inplace=True)
```

```
#Checking for missing values
df_n.isnull().sum()
```

```
df_n.info()
```

```
#Checking for duplicates

df_n.duplicated()
```

```
#dropping the duplicated data
df_n.drop_duplicates()
```

```
#Looking at top n records
df_n.head()
```

```
#Renaming the Value column to Billing price
df_n.rename(columns={'VALUE':'BILLING_PRICE'}, inplace=True)
```

```
#Describe the quick stats (count, mean, std, min, first quartile,median, third quartile, max) on THE PRICE columns

df_n[['BILLING_PRICE']].describe()
```

```
#dropping the unnamed column
df_n.drop
del df_n['Unnamed: 0']
df_n
```
